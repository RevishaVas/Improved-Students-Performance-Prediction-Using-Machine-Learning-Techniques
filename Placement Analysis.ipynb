```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
#reading the file
data=pd.read_csv('CampusPlace.csv')
data
```

```python
#droping the serial no and salary col
data=data.drop('sl_no',axis=1)
data=data.drop('salary',axis=1)
data.head()
```

```python
data.shape
```

```python
#catgorising col for further labelling
data["gender"]=data["gender"].astype('category')
data["ssc_b"]=data["ssc_b"].astype('category')
data["hsc_b"]=data["hsc_b"].astype('category')
data["workex"]=data["workex"].astype('category')
data["specialisation"]=data["specialisation"].astype('category')
data["status"]=data["status"].astype('category')
data["hsc_s"]=data["hsc_s"].astype('category')
data.dtypes
```

```python
#labelling the col
data["gender"]=data["gender"].cat.codes
data["ssc_b"]=data["ssc_b"].cat.codes
data["hsc_b"]=data["hsc_b"].cat.codes
data["workex"]=data["workex"].cat.codes
data["specialisation"]=data["specialisation"].cat.codes
data["status"]=data["status"].cat.codes
data["hsc_s"]=data["hsc_s"].cat.codes
data
```

```python
#selecting the features and labels
X=data.iloc[:,:-1].values
Y=data.iloc[:,-1].values
print(Y.shape)
print(Y)

```

```python
print(X.shape)
print(X)
```

```python
#dividing the data into train and split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_train)
print(X_train.shape)
```

```python
print(X_test)
print(X_test.shape)
```

```python
print(Y_test)
print(Y_test.shape)
```

```python
print(Y_train)
print(Y_train.shape)
```

```python
#creating a classifier using sklearn
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train,Y_train)
#printing the acc
clf.score(X_test,Y_test)
```

```python
#predicting for random value
clf.predict([[1,77,0,95,0,2,80,0,0,1]])
```

```python
clf.predict([[1,57,0,55,0,2,58,0,0,1]])
```

```python

#creating a Y_pred for test data
Y_pred=clf.predict(X_test)
Y_pred
```

```python
#model generation
pickle.dump(clf, open('modelCampus.pkl','wb'))
#evalution of the classifier
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
accuracy_score(Y_test, Y_pred)
```
