```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
#reading the file
data=pd.read_csv('Scholarship.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['caste'].fillna('Other',inplace=True)
```

```python
data
```

```python
data['Scholarship Name'].nunique()
```

```python
data['Scholarship Name'].unique()
```

```python
scholarship_dict={'No Scholarship':0,'Post-Metric':1, 'Karnataka Minority Schlorship':2,
       'Children of Beedi Worker':3, 'Children of Mica/Iron':4,
       'Differently Abled Schlorship':5,
       'Labour welfare commission schlorship':6, 'LIC schlorship':7,
       'NTPC Scholarship Scheme':8, 'VIT University Ignite Scholarships':9,
       'Indian Oil Corporation Scholarships':10}
data['Scholarship Name']=data['Scholarship Name'].map(scholarship_dict)
```

```python
data
```

```python
#droping the serial no and salary col
data=data.drop('Sl No.',axis=1)
data=data.drop('Student Name',axis=1)
data.head()
```

```python
data['gender']=data['gender'].map({'M':0,'F':1})
data
```

```python
data['caste']=data['caste'].map({'Catholic':1,'Muslim':2,'Other':3})
data
```

```python
#selecting the features and labels
X=data.iloc[:,1:].values
Y=data.iloc[:,0].values
print(X.shape)
print(X)
print(Y.shape)
print(Y)
```

```python
#dividing the data into train and split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
print(X_train)
print(X_train.shape)
```

```python
print(Y_train)
print(Y_train.shape)
```

```python
print(X_test.shape)
print(X_test)
```

```python
print(Y_test.shape)
print(Y_test)
```

```python
#creating a classifier using sklearn
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score


lr =  LogisticRegression()
lr.fit(X_train, Y_train)
Y_pred=lr.predict(X_test)

df=pd.DataFrame({'Actual':Y_test,"Predicted":Y_pred})

print(accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))

print(df)
```

```python
#creating a classifier using sklearn
from sklearn.naive_bayes import GaussianNB

nb =  GaussianNB()
nb.fit(X_train, Y_train)
Y_pred=nb.predict(X_test)

df=pd.DataFrame({'Actual':Y_test,"Predicted":Y_pred})

print(accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))

print(df)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, Y_train)
Y_pred=knn.predict(X_test)

df=pd.DataFrame({'Actual':Y_test,"Predicted":Y_pred})

print(accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))

print(df)
```

```python
#model generation
import pickle
pickle.dump(knn, open('scholar_model.pkl','wb'))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_depth=3,random_state=0, max_features = None,  min_samples_leaf = 2)
dtree.fit(X_train, Y_train)
Y_pred=dtree.predict(X_test)

df=pd.DataFrame({'Actual':Y_test,"Predicted":Y_pred})

print(accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))

print(df)

```

```python
from sklearn.svm import SVC

svm =  SVC(gamma='auto')

svm.fit(X_train, Y_train)
Y_pred=svm.predict(X_test)

df=pd.DataFrame({'Actual':Y_test,"Predicted":Y_pred})

print(accuracy_score(Y_test, Y_pred))
print(confusion_matrix(Y_test, Y_pred))

print(df)
```
